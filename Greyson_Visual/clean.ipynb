```python
import pandas as pd
```

```python
df  = pd.read_csv("../data/aac_intakes_outcomes.csv")
```

```python
dog_mask = df.animal_type == 'Dog'
```

```python
dogDf = df[dog_mask].reset_index(drop=True)
```

```python
dogClean = dogDf[['animal_type','breed','intake_datetime','outcome_type','outcome_datetime','time_in_shelter_days']]
```

```python
dogClean2 = dogDf.groupby(['intake_year','intake_month']).intake_datetime.count().reset_index()
dogClean2 = dogClean2.rename(columns={'intake_datetime':'count'})
dogClean2.to_json("data/dog_clean_time.json")
```

```python
dogClean2
```

```python
dogClean.to_csv("data/dog_clean.csv")
dogClean.to_json("data/dog_clean.json")
```
